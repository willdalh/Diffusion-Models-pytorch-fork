```python
import matplotlib.pyplot as plt
import numpy as np

def plot_images(images, title=None, norm=False, norm_across_batch=False, clip=False):
    if norm and clip:
        raise ValueError("Cannot normalize and clip at the same time.")
    
    to_plot = images.detach().clone();
    if norm:
         to_plot = normalize(to_plot, norm_across_batch)
    elif clip:
        to_plot = to_plot.clamp(0, 1)

    n = to_plot.shape[0]
    fig, axs = plt.subplots(1, n, figsize=(n * 3, 3))
    if title:
        fig.suptitle(title, fontsize=20)
    to_plot = to_plot.permute(0, 2, 3, 1)
    for i in range(n):
        curr_axs = axs if n == 1 else axs[i]
        if to_plot.shape[3] == 1:
            curr_axs.imshow(to_plot[i, ..., 0].cpu(), cmap="gray")
        else:
            curr_axs.imshow(to_plot[i].cpu())
           
    plt.show()

def normalize(x, across_batch=False):
    x_copy = x.detach().clone()
    out = None
    if across_batch:
        out = _normalize_across_all_dims(x_copy)
    else:
        for i in range(x_copy.shape[0]):
            x_copy[i] = _normalize_across_all_dims(x_copy[i])
        out = x_copy
    return out
        

def _normalize_across_all_dims(x):
    inv_scale = x.max() - x.min()
    if inv_scale == 0:
        inv_scale = 1
    return (x - x.min()) / (inv_scale)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = UNet().to(device)
print(model.device)
ckpt = torch.load("models/unconditional_ckpt.pt", map_location=device)
model.load_state_dict(ckpt) 
diffusion = Diffusion(img_size=64, device=device)
```

```python
x = diffusion.sample(model, 4)
plot_images(x)
```

## DDIM

```python
n = 4

model.eval()
eta = 0
d = diffusion

def sample_ddim(model, diffusion, eta, x):
    with torch.no_grad():
        for t in range(diffusion.noise_steps-1, 0, -1):
            tt = (torch.ones(n) * t).long().to(device)

            z = torch.randn(x.shape).to(device)

            eps = model(x, tt)

            x0_t = (x - eps * (1 - d.alpha_hat[t]).sqrt()) / d.alpha_hat[t].sqrt()

            c1 = eta * ((1 - d.alpha_hat[t] / d.alpha_hat[t - 1]) * (1 - d.alpha_hat[t - 1]) / (1 - d.alpha_hat[t])).sqrt()

            c2 = ((1 - d.alpha_hat[t - 1]) - c1 ** 2).sqrt()

            x = d.alpha_hat[t - 1].sqrt() * x0_t + c1 * z + c2 * eps
    return x

#         print(d.alpha_hat[t] * torch.randn(3, 3).to(device))
```

```python
eta = 1
x = torch.randn(n, 3, 64, 64).to(device)
x = sample_ddim(model, diffusion, eta, x)
plot_images(x, norm=False, clip=True)
print(x.amax(dim=(1, 2, 3)))
print(x.amin(dim=(1, 2, 3)))
```

```python
eta = 0
x = torch.randn(n, 3, 64, 64).to(device)
x = sample_ddim(model, diffusion, eta, x)
plot_images(x, norm=False, clip=True)
print(x.amax(dim=(1, 2, 3)))
print(x.amin(dim=(1, 2, 3)))
```
